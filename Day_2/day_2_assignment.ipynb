Lets work on this assignment


```python
class Book:
    def __init__(self, book_id, book_name):
        # Book ID and Book Name
        self.book_id = book_id
        self.book_name = book_name
        self.taken_by = None  # Initially, no student has taken the book

    def __repr__(self):
        return f"Book({self.book_id}, {self.book_name})"


class Student:
    def __init__(self, roll_no, name, faculty):
        # Roll No, Name and Faculty
        self.roll_no = roll_no
        self.name = name
        self.faculty = faculty
        self.books_taken = []

    def display_student(self):
        print(f'Roll No: {self.roll_no}\nName: {self.name}\nFaculty: {self.faculty}')


class LibraryManagementSystem:
    def __init__(self):
        self.books = []
        self.students = []

    def add_book(self, book):
        self.books.append(book)

    def add_student(self, student):
        self.students.append(student)

    def assign_book(self, student, book):
        """
        Assign a book to a student
        """
        if book in self.books and book.taken_by is None:
            student.books_taken.append(book)
            book.taken_by = student
        else:
            print(f"Book '{book.book_name}' is not available in the library.")

    def display_books_taken_by_students(self):
        """
        Print the books taken by each student
        """
        for student in self.students:
            print(f"Books taken by {student.name}:")
            for book in student.books_taken:
                print(f"- {book.book_name}")

    def display_books_taken_from_library(self):
        """
        Display books taken from library along with the student name
        """
        for book in self.books:
            if book.taken_by is not None:
                print(f"Book '{book.book_name}' taken by {book.taken_by.name}")


# Example usage
# Create instances of Book and Student
book1 = Book(1, "OS")
book2 = Book(2, "DBMS")
book3 = Book(3, "DSA")

student1 = Student(101, "Alice", "BEI")
student2 = Student(102, "Bob", "BCT")
student3 = Student(103, "Charlie", "BCE")

# Create Library Management System
library_system = LibraryManagementSystem()
library_system.add_book(book1)
library_system.add_book(book2)
library_system.add_student(student1)
library_system.add_student(student2)

# Assign books to students
library_system.assign_book(student1, book1)
library_system.assign_book(student2, book2)
library_system.assign_book(student3, book1)

# Display books taken by students and books taken from library
library_system.display_books_taken_by_students()
library_system.display_books_taken_from_library()

```
